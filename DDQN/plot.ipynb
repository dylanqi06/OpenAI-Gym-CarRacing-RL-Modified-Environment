```python
import csv
import numpy as np
import matplotlib.pyplot as plt

# Parameters
csv_filename = 'rewards/dylanqi/DDQN2/20241123-005953/episode_1950.csv'  # Replace with your actual filename
window_size = 100  # Number of episodes to average over
rewards = []
with open(csv_filename, 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        if len(row) > 0:
            # The first value in each row is the reward
            reward = float(row[0])
            rewards.append(reward)
```

```python
running_avg = np.empty(len(rewards))
for i in range(len(rewards)):
    start_index = max(0, i - window_size + 1)
    running_avg[i] = np.mean(rewards[start_index:i+1])

# Plot the running average
plt.figure(figsize=(10,6))
plt.plot(running_avg)
plt.title("Running average of previous {} rewards".format(window_size))
plt.xlabel("Episode")
plt.ylabel("Average Reward")
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

# Replace with your actual CSV filename
csv_filename = 'episode_test_runs/oah33/20241127-013421/DDQN2/episode_run_rewards.csv'

episodes = []
averages = []

# Read the CSV and extract episode number and average of 10 runs
with open(csv_filename, 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        if len(row) > 1:
            episode = float(row[0])
            avg_10_runs = float(row[1])
            episodes.append(episode)
            averages.append(avg_10_runs)

# Plotting
plt.figure(figsize=(10,6))
plt.plot(episodes, averages, marker='o', linestyle='-')
plt.title('Average of 10 Runs per Evaluation')
plt.xlabel('Episode')
plt.ylabel('Average Evaluation Score')
plt.show()
```
