```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam

# Define the architecture
model = Sequential()
model.add(Conv2D(filters=6, kernel_size=(7, 7), strides=3, activation='relu', input_shape=(96, 96, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(filters=12, kernel_size=(4, 4), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(216, activation='relu'))
model.add(Dense(12, activation=None))  # Update with the number of actions in your action space

# Compile the model (necessary to initialize it)
model.compile(loss='mean_squared_error', optimizer=Adam(learning_rate=0.001, epsilon=1e-7))

# Load the weights
model.load_weights('model/dylanqi/DDQN2/20241122-163152/episode_100.weights.h5')

# Print the layers and weights
for layer in model.layers:
    print(f"Layer: {layer.name}")
    weights = layer.get_weights()
    print(f"Weights shape: {[w.shape for w in weights]}")
    print("Weights:", weights)
```
